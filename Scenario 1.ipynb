```python
import pandas as pd

retaildf = pd.read_csv('/retail_sales.csv')
display(retaildf.head())
```

```python
retaildf.info()
```

```python
retaildf.describe()
```

```python
# To Find Duplicates
duplicates = retaildf[retaildf.duplicated()]
print(duplicates)
```

```python
dropdupdf = retaildf.drop_duplicates()
print(dropdupdf)
```

```python
dropdupdf.isnull().sum()
```

```python
# Handling missing values
# using mean of to place missing values as it is a numerical data based on Category
dropdupdf['Price'] = dropdupdf['Price'].fillna(dropdupdf.groupby("Category")['Price'].transform('mean'))
print(dropdupdf)
```

```python
dropdupdf['Date'] = pd.to_datetime(dropdupdf['Date'], dayfirst=True).dt.strftime('%Y-%m-%d')
print(dropdupdf)
```

```python
# Filter Data with only first word in category column
dropdupdf['Category'] = dropdupdf['Category'].str.split().str[0]
print(dropdupdf['Category'])
```

```python
# Categorize products based on price
dropdupdf['Price_Category'] = dropdupdf['Price'].apply(lambda x: 'Expensive' if x > 500 else 'Affordable')

print(dropdupdf)
```
